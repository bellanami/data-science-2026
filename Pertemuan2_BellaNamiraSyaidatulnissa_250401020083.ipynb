```python
# =================================================================
# PERTEMUAN 2: STRUKTUR DATA PYTHON, NUMPY & PANDAS
# Nama: Bella Namira Syaidatulnissa
# NIM:  250401020083
# =================================================================

# 4. Impor library yang dibutuhkan [cite: 312, 321]
import pandas as pd
import numpy as np

# 5. Muat dataset Titanic dari URL GitHub [cite: 322, 323]
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# --- JAWABAN PERTANYAAN LATIHAN (Q1-Q6) --- [cite: 326]

# Q1: Berapa jumlah baris dan kolom? Kolom apa saja yang tersedia? [cite: 326, 327]
print("--- Q1 ---")
print(f"Jumlah baris dan kolom: {df.shape}") # [cite: 230]
print(f"Daftar kolom: {df.columns.tolist()}\n") # [cite: 234]

# Q2: Kolom mana yang memiliki missing values? Jumlah dan proporsinya? [cite: 327, 328]
print("--- Q2 ---")
missing_data = df.isnull().sum()
total_rows = len(df)
missing_info = pd.DataFrame({
    'Jumlah Missing': missing_data,
    'Proporsi (%)': (missing_data / total_rows) * 100
})
print(missing_info[missing_info['Jumlah Missing'] > 0], "\n")

# Q3: Berapa persen penumpang yang selamat (Survived=1)? [cite: 328, 329]
print("--- Q3 ---")
# Menggunakan Pandas mean() karena kolom Survived berisi 0 dan 1 [cite: 289]
percent_survived = df['Survived'].mean() * 100
print(f"Persentase penumpang selamat: {percent_survived:.2f}%\n")

# Q4: Filter penumpang wanita dari kelas 1 (Pclass=1). Jumlah dan rata-rata usia? [cite: 329, 330]
print("--- Q4 ---")
# Menggunakan Boolean Indexing dengan kondisi ganda [cite: 274, 275]
wanita_kelas1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]
print(f"Jumlah penumpang wanita kelas 1: {len(wanita_kelas1)}")
print(f"Rata-rata usia: {wanita_kelas1['Age'].mean():.2f} tahun\n")

# Q5: Hitung tingkat keselamatan (mean Survived) per kelas penumpang. [cite: 330, 331]
print("--- Q5 ---")
# Menggunakan groupby dan agregasi mean [cite: 281, 289]
survival_by_class = df.groupby('Pclass')['Survived'].mean()
print(survival_by_class)
print(f"Kelas dengan tingkat keselamatan tertinggi: Kelas {survival_by_class.idxmax()}\n")

# Q6: Analisis tambahan: Hubungan antara jenis kelamin dan keselamatan [cite: 331]
print("--- Q6 ---")
# Analisis tambahan menggunakan groupby multi-kolom [cite: 304, 305]
analisis_gender = df.groupby('Sex')['Survived'].mean()
print("Tingkat keselamatan berdasarkan jenis kelamin:")
print(analisis_gender)
```
